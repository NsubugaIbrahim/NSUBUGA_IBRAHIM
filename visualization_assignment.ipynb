```python
#Data Preprocessing or Data Cleaning
#Bad data handling
#1.empty values
#2.wrong formats    30/03/03     300303    30-03-03
#3.wrong data     Age 23    -23  123
#4.duplicates    more entries from the same person
import pandas as pd
import numpy as np
import matplotlib as plt
df = pd.read_csv('data.csv')
#check for empty cells
print(df.isnull().sum())

#check wrong formats
print(df.dtypes)
```

```python
#wrong data
#check for negative values in numeric columns
for col in df.select_dtypes(include=[np.number]).columns:
    if (df[col] < 0).any():
        print(f"Negative values found in column: {col}")
```

```python
#Duplicates
duplicates = df.duplicated()
if duplicates.any():
    print("Duplicates found:")
    print(df[duplicates])
```

```python
#recommendations
#1. convert to correct format
#2. fill empty values with mean for numeric values
#3. remove duplicates in rows not in columns
#4. correct wrong data

#EDA Exploratory Data Analysis
#check the shape of the data
print("Shape of the data:", df.shape)
```

```python
#fill empty values with mean for numeric columns
for col in ['Duration', 'Pulse','Maxpulse', 'Calories']:
    df[col].fillna(df[col].mean(), inplace=True)
```

```python
#Data visualization......Quickest way to make sense out of data
#Libraries used for data visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Histogram of 'Duration'
# Set up the figure size
plt.figure(figsize=(15, 10))

# Histogram of 'Duration'
plt.subplot(2, 2, 1)
sns.histplot(df['Duration'], bins=30, kde=True)
plt.title('Duration Distribution')
# Histogram of 'Pulse'
plt.subplot(2, 2, 2)
sns.histplot(df['Pulse'], bins=30, kde=True)
plt.title('Pulse Distribution')

# Exercise Plot for Maxpulse and Colories for Histrogram
```

```python
# Scatter plot of 'Duration' vs 'Pulse'
plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 1)
sns.scatterplot(x='Duration', y='Pulse', data=df)
plt.title('Duration vs Pulse')
# Scatter plot of 'Pulse'
plt.subplot(2, 2, 2)
sns.scatterplot(x='Pulse', y='Calories', data=df)
plt.title('Pulse vs Calories')
```

```python
# Correlation heatmap
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
